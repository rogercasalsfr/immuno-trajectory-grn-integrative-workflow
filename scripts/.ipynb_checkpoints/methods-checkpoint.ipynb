In here to perform trajectory analysis, we should subset the data based on previous biology and our interests.

There are different types of approaches, we know that single-cell RNA seq is a snapchot for a particular moment, so in this case we also filter and see how is the trajectory before the treatment to then further compare with after the treatment

```python
# we need to subset for just one cell type CD4 or CD8
CD4 = ['Naive', 'Tfh', 'Th17', 'Treg']
adata = adata[adata.obs['cluster'].isin(CD4)] 
adata = adata[adata.obs['treatment'] == 'pre'] # column that has "pre" and "post" treatment information


# subset adata for cleaner easy analysis
np.random.seed(42)  # reproducibility
cells = np.random.choice(adata.obs_names, size=3000, replace=False)
adata_sub = adata[cells].copy()

# We do get 3k cells for easy downstream analysis tasks.
adata = adata_sub.copy()
```

# Methods that we will use

A key step in trajectory analysis is choosing the method used to infer cell state transitions.  
Several algorithms have been developed to reconstruct trajectories from single-cell RNA-seq data.  
A useful review comparing many of these methods can be found here:  
https://www.nature.com/articles/s41587-019-0071-9

In this project we will explore three commonly used trajectory inference methods.


## Monocle3

Monocle3 reconstructs trajectories by learning a graph that connects cells in a low-dimensional space (for example UMAP).  
Cells are then ordered along this trajectory using **pseudotime**, which represents the progression of a biological process such as differentiation.

Monocle3 is widely used and works well for identifying **branching trajectories**.


## Slingshot

Slingshot infers trajectories starting from **clusters of cells**.  
It first connects clusters using a **minimum spanning tree** and then fits smooth curves that represent the possible developmental paths between cell states.

This method is simple, robust, and works well when clusters are clearly defined.


## PAGA

PAGA (Partition-based Graph Abstraction) summarizes how **clusters of cells are connected**.  
Instead of focusing on individual cells, it builds a graph that represents the **global structure of cell transitions**.

PAGA is particularly useful for understanding the overall topology of the trajectory and validating lineage relationships.
